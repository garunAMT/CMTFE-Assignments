# **Question 3**

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining function
def F(x):
    return x**2 + 2*x + 5

def analytical_F_prime(x):
    return 2*x + 2

# domain length
L = 1

def numerical_derivative(x, dx, scheme='forward'):
    N = len(x)
    numerical_F_prime = np.zeros(N)

    for i in range(1, N-1):
        if scheme == 'forward':
            numerical_F_prime[i] = (F(x[i+1]) - F(x[i])) / dx
        elif scheme == 'backward':
            numerical_F_prime[i] = (F(x[i]) - F(x[i-1])) / dx
        elif scheme == 'central':
            numerical_F_prime[i] = (F(x[i+1]) - F(x[i-1])) / (2 * dx)

    # Handle boundary cases
    if scheme == 'forward':
        numerical_F_prime[-1] = (F(x[-1]) - F(x[-2])) / dx
    elif scheme == 'backward':
        numerical_F_prime[0] = (F(x[1]) - F(x[0])) / dx

    return numerical_F_prime

# Plotting function
def plot_solutions(x, analytical_F_prime, numerical_F_prime, title):
    plt.plot(x, analytical_F_prime, label='Analytical F\'(x)', color='blue')
    plt.plot(x, numerical_F_prime, label='Numerical F\'(x)', color='red', linestyle='--')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('F\'(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Case 1: Forward Difference with dx = 0.1
dx = 0.1
x = np.linspace(0, L, int(L/dx) + 1)
numerical_F_prime = numerical_derivative(x, dx, scheme='forward')
analytical_F_prime_values = analytical_F_prime(x)
plot_solutions(x, analytical_F_prime_values, numerical_F_prime, 'Forward Difference Scheme (dx = 0.1)')

# Case 2: Backward Difference with dx = 0.02
dx = 0.02
x = np.linspace(0, L, int(L/dx) + 1)
numerical_F_prime = numerical_derivative(x, dx, scheme='backward')
analytical_F_prime_values = analytical_F_prime(x)
plot_solutions(x, analytical_F_prime_values, numerical_F_prime, 'Backward Difference Scheme (dx = 0.02)')

# Case 3: Central Difference with dx = 0.01
dx = 0.01
x = np.linspace(0, L, int(L/dx) + 1)
numerical_F_prime = numerical_derivative(x, dx, scheme='central')
analytical_F_prime_values = analytical_F_prime(x)
plot_solutions(x, analytical_F_prime_values, numerical_F_prime, 'Central Difference Scheme (dx = 0.01)')
```

```python
# question 2
import numpy as np
import matplotlib.pyplot as plt

# Defining the function and its second derivative
def F(x):
    return 2*x**4 + 3*x**3 + 4*x**2 + 5*x + 6

def analytical_F_double_prime(x):
    return 24*x**2 + 18*x + 8

# Numerical second derivative using Central Difference Scheme
def numerical_second_derivative(x, dx):
    N = len(x)
    numerical_F_double_prime = np.zeros(N)

    for i in range(1, N-1):
        numerical_F_double_prime[i] = (F(x[i+1]) - 2*F(x[i]) + F(x[i-1])) / (dx**2)

    return numerical_F_double_prime

# Defining the domain and dx
dx = 0.5
x = np.linspace(0, 10, int(10/dx) + 1)

# Calculation of the numerical and analytical second derivatives
numerical_F_double_prime = numerical_second_derivative(x, dx)
analytical_F_double_prime_values = analytical_F_double_prime(x)

# Plot between the 2nd node and the penultimate node
plt.plot(x[1:-1], analytical_F_double_prime_values[1:-1], label='Analytical F\'\'(x)', color='blue')
plt.plot(x[1:-1], numerical_F_double_prime[1:-1], label='Numerical F\'\'(x)', color='red', linestyle='--')
plt.title('Second Derivative: Analytical vs Numerical (Central Difference, dx = 0.5)')
plt.xlabel('x')
plt.ylabel('F\'\'(x)')
plt.legend()
plt.grid(True)
plt.show()

# Justification of the solution
error = np.abs(analytical_F_double_prime_values[1:-1] - numerical_F_double_prime[1:-1])
print(f"Max error between analytical and numerical solution: {np.max(error)}")
```

# Question 5


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 5*x + 6

# Analytical first derivative
def analytical_derivative(x):
    return 5 * np.ones_like(x)

# Central difference scheme for the first derivative
def central_difference(f, x, dx):
    return (f(x + dx) - f(x - dx)) / (2 * dx)

# Parameters for the problem
dx = 0.01
x_domain = np.arange(0, 1 + dx, dx)  # Domain [0, 1] with step size dx

# Compute the analytical derivative
analytical_f_prime = analytical_derivative(x_domain)

# Compute the numerical derivative using central difference
numerical_f_prime = np.zeros_like(x_domain)
n = len(x_domain)
for i in range(1, n-1):
    numerical_f_prime[i] = central_difference(f, x_domain[i], dx)

# Handling boundaries (not central difference here)
numerical_f_prime[0] = (f(x_domain[1]) - f(x_domain[0])) / dx  # Forward difference at first node
numerical_f_prime[-1] = (f(x_domain[-1]) - f(x_domain[-2])) / dx  # Backward difference at last node

# Plotting results between the 2nd node and the penultimate node
plt.plot(x_domain[1:-1], numerical_f_prime[1:-1], label='Numerical Derivative')
plt.plot(x_domain[1:-1], analytical_f_prime[1:-1], label='Analytical Derivative', linestyle='--')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Analytical vs Numerical First Derivative")
plt.legend()
plt.show()

# Check if the solutions are the same
difference = np.abs(analytical_f_prime[1:-1] - numerical_f_prime[1:-1])
print("Max difference between numerical and analytical solution:", np.max(difference))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.sin(x) + np.exp(1-x)

# Analytical second derivative
def analytical_second_derivative(x):
    return -np.sin(x) + np.exp(1-x)

# Central difference scheme for the second derivative
def central_difference_second_derivative(f, x, dx):
    return (f(x + dx) - 2*f(x) + f(x - dx)) / (dx**2)

# Parameters for the problem
dx = 0.01
x_domain = np.arange(0, 1 + dx, dx)  # Domain [0, 1] with step size dx

# Compute the analytical second derivative
analytical_f_double_prime = analytical_second_derivative(x_domain)

# Compute the numerical second derivative using central difference
numerical_f_double_prime = np.zeros_like(x_domain)
n = len(x_domain)
for i in range(1, n-1):
    numerical_f_double_prime[i] = central_difference_second_derivative(f, x_domain[i], dx)

# Handling boundaries (not central difference here)
numerical_f_double_prime[0] = (f(x_domain[1]) - 2*f(x_domain[0]) + f(x_domain[0] - dx)) / (dx**2)
numerical_f_double_prime[-1] = (f(x_domain[-1] + dx) - 2*f(x_domain[-1]) + f(x_domain[-2])) / (dx**2)

# Plotting results between the 2nd node and the penultimate node
plt.plot(x_domain[1:-1], numerical_f_double_prime[1:-1], label='Numerical Second Derivative')
plt.plot(x_domain[1:-1], analytical_f_double_prime[1:-1], label='Analytical Second Derivative', linestyle='--')
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.title("Analytical vs Numerical Second Derivative")
plt.legend()
plt.show()

# Check if the solutions are the same
difference = np.abs(analytical_f_double_prime[1:-1] - numerical_f_double_prime[1:-1])
print("Max difference between numerical and analytical solution:", np.max(difference))
```
